```python
def import_data(filename):
    with open(filename,'r') as f:
        
        X = []
        y = []
        
        for line in f:
            line = line.replace("?","nan").strip().split(',')
            
            floatList = [float(i) for i in line]

            tail = floatList.pop(-1)
            head = floatList           

            X.append(head)
            
            y.append(tail)          
            y = [int(i) for i in y]

    return X, y

print(import_data('arrhythmia.data')[0])
print('\n')
print(import_data('arrhythmia.data')[1])
```

```python
#2a

def median(lst):
    
    length = len(lst)
    sort = sorted(lst)
    
    return ((sort[(length // 2)] + sort[~length // 2]) / 2)

def addmedian(lst,median):
        
    emptyList = []
    for i in range(len(lst)):
        if (math.isnan(lst[i]) != True):
            emptyList.append(lst[i])
        else:
            emptyList.append(median)
    listWithMedian = emptyList
    
    return listWithMedian

def inmute_missing(X):
    
    transposed = list(map(list, zip(*X)))
    
    emptyList = [[]]*len(transposed)
    emptyList2 = [[]]*len(transposed)
    emptyList3 = [[]]*len(transposed)
    
    for i in range(len(transposed)):
        emptyList[i] = [j for j in transposed[i] if math.isnan(j) != True]
    noNanList = emptyList
    
    for i in range(len(transposed)):
        emptyList3[i] = [j for j in X[i] if math.isnan(j) != True]
    
    medianList = []
    
    for i in noNanList:
        medianList.append(median(i))
    
    
    for j in X:
        for i in range(len(medianList)):
            emptyList2[i] = addmedian(j, medianList[i])
        X = emptyList2
        
        
        return X
    
X1 = (import_data('arrhythmia.data')[0])
y1 = (import_data('arrhythmia.data')[1])

print(inmute_missing(X1))

X2 = (inmute_missing(X1))
```

Median is better because it avoids outliers that are small in numbers but are so radically different that they influence the mean. By taking median we avoid it and go straight to the middle, where numbers that are most often popping are located, making it much more accurate and almost improbable to be skewed by outliers

```python
#2c

def discard_missing(X,y):
    
    emptyList = [[]]*len(X)
    for i in range(len(X)):
        emptyList[i] = [j for j in X[i] if math.isnan(j) != True]
    X = emptyList
    
    yList = [j for j in y if math.isnan(j) != True]
    y = yList
    
    return X, y

print(discard_missing(X1,y1)[0])
print('\n')
print(discard_missing(X1,y1)[1])
```

```python
#3a

def shuffle_data(X, y):
    
    emptyList = []
    emptyList2 = []
    orderedList = []
    
    for i in range(len(X)):
        orderedList.append(i)

    shuffledList = np.array(orderedList)
    
    np.random.shuffle(shuffledList)

    for i in shuffledList:
        emptyList.append(X[i])
        emptyList2.append(y[i])
    X = emptyList
    y = emptyList2
    
    return X, y

X2, y2 = discard_missing(X1,y1)

print(shuffle_data(X2,y2)[0])
print('\n')
print(shuffle_data(X2,y2)[1])
```

```python
#3b

def compute_std(X):
    std = []
    transposed = list(map(list, zip(*X)))
    for i in transposed:
        mean = sum(i) / len(i)
        stdSingle = math.sqrt(sum((j-(sum(i) / len(i)))**2 for j in i) / (len(i)))
        std.append(stdSingle)
    return std

print (compute_std(X2))
```

```python
#3c

def remove_outlier(X, y):
    
    std = []
    mean = []
    
    transposed = list(map(list, zip(*X)))
    for i in transposed:
        mean.append(sum(i) / len(i))
        stdSingle = math.sqrt(sum((j-(sum(i) / len(i)))**2 for j in i) / (len(i)))
        std.append(stdSingle)
        
    outlierFreeList = []
    
    j = 0
    for i in transposed:
        outliers = []
        lowCut = mean[j] - std[j]*2
        highCut = mean[j] + std[j]*2
        outliersGone = [x == 'nan' if (x < lowCut or x > highCut) else x for x in i]
        outlierFreeList.append(outliersGone)               
        j = j+1
    
    backTranspose = list(map(list, zip(*outlierFreeList)))
    
    X = []
    j = 0
    for i in backTranspose:
        part = [x for x in i if x != False]
        X.append(part)
        
    for i in range(len(y)):
        outliers = []
        lowCut = (sum(y) / len(y)) - (math.sqrt(sum((j-(sum(y) / len(y)))**2 for j in y) / (len(y))))*2
        highCut = (sum(y) / len(y)) + (math.sqrt(sum((j-(sum(y) / len(y)))**2 for j in y) / (len(y))))*2
        outliersGone2 = [x == 'nan' if (x < lowCut or x > highCut) else x for x in y]
    
    y = []
    for i in outliersGone2:
        y = [x for x in outliersGone2 if x != False]
    
    return X, y

print(remove_outlier(X2, y1))

X3 = remove_outlier(X2, y1)[0]
Y3 = remove_outlier(X2, y1)[1]
```

```python
#3d 

def standardize_data(X):
    
    std = []
    mean = []
    
    transposed = list(map(list, zip(*X)))
    for i in transposed:
        mean.append(sum(i) / len(i))
        stdSingle = math.sqrt(sum((j-(sum(i) / len(i)))**2 for j in i) / (len(i)))
        std.append(stdSingle)
        
    transposed2 = list(map(list, zip(*X)))
    
    for i in range(len(transposed2)):
        k = 0
        for j in transposed2[i]:
            if std[i] == 0:
                transposed2[i][k] = j
            else:
                transposed2[i][k] = ((j-mean[i])/std[i])
            k = k+1
    
    return transposed2

#time and space complexity here is O(m,n) since they (m and n) are the dimensions of the data set

print(standardize_data(X3))
```

```python
def import_data2(filename):
    with open(filename,'r') as f:
        headers = f.readline()
        
        X = [[] for i in range(8)]
        y = []
        
        # REFERENCE:
        #
        # x[0] - PassengerId
        # y - Survived
        # x[1] - Pclass
        # nan - Name
        # x[2] - Sex
        # x[3] - Age
        # x[4] - SibSp
        # x[5] - Parch
        # nan - Ticket
        # x[6] - Fare
        # nan - Cabin
        # x[7] - Embarked
        
        info = f.read().strip().split('\n')
        for line in info:
            line = line.strip().split(',')
            
            X[0].append(int(line[0]))
            y.append(int(line[1]))
                        
            X[1].append(int(line[2]))
            #Name1.append(line[3])
            #Name2.append(line[4])
            X[2].append(line[5])
            if line[6] == '':
                X[3].append(float('nan'))
            else:
                X[3].append(float(line[6]))
            X[4].append(int(line[7]))
            X[5].append(int(line[8]))
            #Ticket.append(int(line[9]))
            X[6].append(float(line[10]))
            #Cabin.append(int(line[11]))
            if line[12] == '':
                X[7].append(float('nan'))
            else:
                X[7].append(str(line[12]))      
           
        #female - 0
        #male - 1
        
        j = 0
        for i in X[2]:
            if i == 'male':
                X[2][j] = 1
            else:
                X[2][j] = 0
            j = j+1
        
        #C - 0
        #Q - 1
        #S - 2
        
        j = 0
        for i in X[7]:
            if i == 'C':
                X[7][j] = 0
            elif i == 'Q':
                X[7][j] = 1
            else:
                X[7][j] = 2
            j = j+1
            
        return X, y

print(import_data2('train.csv'))
```

```python
#5a

def train_test_split(X, y, t_f):
    
    X_test, X_train = np.split(shuffle_data(X,y)[0],[int(t_f * len(X))]) 
    y_test, y_train = np.split(shuffle_data(X,y)[1],[int(t_f * len(y))]) 
    
    return X_train, y_train, X_test, y_test

print(train_test_split(X1, y1, 0.4)[0])
print('\n')
print(train_test_split(X1, y1, 0.4)[1])
print('\n')
print(train_test_split(X1, y1, 0.4)[2])
print('\n')
print(train_test_split(X1, y1, 0.4)[3])
```

```python
#5b

def train_test_CV_split(X, y, t_f, cv_f):
    
    X_test, X_train = np.split(shuffle_data(X,y)[0],[int(t_f*len(X))]) 
    y_test, y_train = np.split(shuffle_data(X,y)[1],[int(t_f*len(y))]) 
    
    X_train, X_cv = np.split(X_train,[int(cv_f*len(X_train))])
    y_train, y_cv = np.split(y_train,[int(cv_f*len(y_train))])
    
    return X_train, y_train, X_test, y_test, X_cv, y_cv

print(train_test_CV_split(X1, y1, 0.4, 0.3)[0])
print('\n')
print(train_test_CV_split(X1, y1, 0.4, 0.3)[1])
print('\n')
print(train_test_CV_split(X1, y1, 0.4, 0.3)[2])
print('\n')
print(train_test_CV_split(X1, y1, 0.4, 0.3)[3])
print('\n')
print(train_test_CV_split(X1, y1, 0.4, 0.3)[4])
print('\n')
print(train_test_CV_split(X1, y1, 0.4, 0.3)[5])
```
